```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Tema dashboard
sns.set_style("white")
plt.rcParams.update({
    "axes.titlesize": 14,
    "axes.labelsize": 12,
    "axes.edgecolor": "gray",
    "axes.linewidth": 0.8,
    "figure.facecolor": "#f7f7f7",
    "axes.facecolor": "white",
    "grid.color": "#e0e0e0",
    "font.family": "sans-serif"
})

# Data fiktif
data = {
    'Negara': ['AS', 'Tiongkok', 'ASEAN', 'Eropa', 'Jepang'],
    'Ekspor_Indonesia': [50, 70, 120, 40, 30],
    'Pertumbuhan_Ekonomi': [2.0, 3.0, 4.5, 1.5, 1.0],
    'Aliran_Modal': [-20, -15, 10, 25, 30]
}
df = pd.DataFrame(data)

# Palet merah
red_palette = sns.color_palette("Reds", len(df))

# Membuat dashboard layout
fig, axs = plt.subplots(2, 2, figsize=(18, 10))
fig.suptitle('📊 Dashboard Ekonomi Indonesia 2025', fontsize=22, fontweight='bold', y=1.05, color='firebrick')

# Panel 1: Ekspor Indonesia
sns.barplot(x='Ekspor_Indonesia', y='Negara', data=df, palette=red_palette, ax=axs[0, 0])
axs[0, 0].set_title('Ekspor Indonesia per Negara', fontweight='bold', color='firebrick')
axs[0, 0].set_xlabel('Ekspor (Miliar USD)')
axs[0, 0].set_ylabel('')
axs[0, 0].grid(True, which='major', axis='x', linestyle='--', alpha=0.7)

# Panel 2: Pertumbuhan Ekonomi
sns.lineplot(x='Negara', y='Pertumbuhan_Ekonomi', data=df, marker='o', markersize=8, linewidth=2, color='firebrick', ax=axs[0, 1])
axs[0, 1].set_title('Pertumbuhan Ekonomi per Negara', fontweight='bold', color='firebrick')
axs[0, 1].set_ylabel('Pertumbuhan (%)')
axs[0, 1].set_xlabel('')
axs[0, 1].grid(True, which='major', axis='y', linestyle='--', alpha=0.7)

# Panel 3: Aliran Modal
sns.barplot(x='Aliran_Modal', y='Negara', data=df, palette=red_palette[::-1], ax=axs[1, 0])
axs[1, 0].set_title('Aliran Modal Global', fontweight='bold', color='firebrick')
axs[1, 0].set_xlabel('Modal Masuk/Keluar (Miliar USD)')
axs[1, 0].set_ylabel('')
axs[1, 0].grid(True, which='major', axis='x', linestyle='--', alpha=0.7)

# Panel 4: Bubble Chart Ekspor vs Modal
bubble = axs[1, 1].scatter(df['Ekspor_Indonesia'], df['Aliran_Modal'],
                           s=df['Pertumbuhan_Ekonomi'] * 80, alpha=0.8, color='indianred', edgecolors='gray', linewidth=1.5)

for i in range(len(df)):
    axs[1, 1].text(df['Ekspor_Indonesia'][i]+1, df['Aliran_Modal'][i], df['Negara'][i], fontsize=10)

axs[1, 1].set_title('Ekspor vs Modal (Ukuran = Pertumbuhan)', fontweight='bold', color='firebrick')
axs[1, 1].set_xlabel('Ekspor (Miliar USD)')
axs[1, 1].set_ylabel('Aliran Modal (Miliar USD)')
axs[1, 1].grid(True, linestyle='--', alpha=0.7)

# Tata letak rapi
plt.tight_layout(pad=3)
plt.subplots_adjust(top=0.88)
plt.show()
```
